```python
import re

text = """
John has 5 apples.
Mary has 12 bananas.
Alice has 1 orange.
Email me at test@example.com.
"""
```

Wildcard .

Matches any single character.

```python
pattern = r"J.hn"
matches = re.findall(pattern, text)
print(matches)  # Output: ['John']
```



Character Classes [ ]

Matches any one character inside the brackets.



```python
pattern = r"[JM]ary"
matches = re.findall(pattern, text)
print(matches)  # Output: ['Mary']
```

[JM] matches either J or M.

Negation [^ ]

```python
pattern = r"[^a-zA-Z\s]+"
matches = re.findall(pattern, text)
print(matches)  # Output: ['5', '12', '1', '.', '.', '.']
```

Matches all non-letter, non-space characters (numbers, punctuation).

3️⃣ Shorthand Character Classes

\d → digits, \w → alphanumeric, \s → whitespace

```python
digits = re.findall(r"\d+", text)
print(digits)  # Output: ['5', '12', '1']

words = re.findall(r"\w+", text)
print(words)  # Output: ['John', 'has', '5', 'apples', 'Mary', ...]
```

4️⃣ Quantifiers

+ → 1 or more, * → 0 or more, ? → 0 or 1

```python
pattern = r"\d+"  # one or more digits
print(re.findall(pattern, text))  # ['5', '12', '1']

pattern = r"a*"  # zero or more 'a'
print(re.findall(pattern, "aaaab"))  # ['aaaa', '', '', '', '']
```

5️⃣ Alternation |

Matches one option or another.

```python
pattern = r"John|Mary"
matches = re.findall(pattern, text)
print(matches)  # ['John', 'Mary']
```

6️⃣ Grouping ( )

Groups patterns for capturing or alternation.

```python
pattern = r"(\w+) has (\d+) (\w+)"
matches = re.findall(pattern, text)
print(matches)
# Output: [('John', '5', 'apples'), ('Mary', '12', 'bananas'), ('Alice', '1', 'orange')]
```

(\w+) captures the name, number, and fruit separately.

7️⃣ Anchors ^ and $

^ → start of string, $ → end of string

```python
pattern = r"^John"
matches = re.findall(pattern, text, re.MULTILINE)
print(matches)  # ['John']

pattern = r"com\.$"
matches = re.findall(pattern, text, re.MULTILINE)
print(matches)  # ['com.']
```

8️⃣ Modifiers

re.I → case-insensitive, re.M → multi-line

```python
pattern = r"john"
matches = re.findall(pattern, text, re.I)
print(matches)  # ['John']
```

9️⃣ Email Example

```python
pattern = r"[\w\.-]+@[\w\.-]+"
emails = re.findall(pattern, text)
print(emails)  # ['test@example.com']
```

[\w\.-]+ → matches letters, digits, . or - one or more times.

@ → matches literal @.

[\w\.-]+ → matches domain part.
